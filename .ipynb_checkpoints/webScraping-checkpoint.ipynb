# Web scraping for product information

This is a web scraping example from [here](https://medium.com/@wahyudihandry/how-to-build-web-scraping-using-beautifulsoup-and-flask-part-i-ca38a167c236).


Go to https://www.bukalapak.com/ and search popular. 
A bunch of lovely megazines will show up.

Inspect html with Google Developer Tool.
One will find out how the product information is stored:
1. All products are in `<div class=”product-card”>`
2. Product name and link are in `<a class=”product__name”>`
3. Product picture in `<img class=”product-media__img”>`
4. Product price in `<span class=”amount”>`
5. Product reviews in `<a class=”review__aggregate”>`


Now start to code.

```python
from bs4 import BeautifulSoup
import requests

# def scrape (
# ...
# )

base_url = "https://www.bukalapak.com/products?utf8=%E2%9C%93&source=navbar&from=omnisearch&search_source=omnisearch_keyword&search%5Bhashtag%5D=&search%5Bkeywords%5D=majalah+popular"

r = requests.get(base_url)
soup = BeautifulSoup(r.text, "html.parser")
all_product = soup.find_all('div', class_="product-card")
l = []
```

Now fill list `l` with products. Each product is a dictionary.

```python
for item in all_product:
    d = { }
    
    product_name = item.find("a", {"class":"product__name"}) 
    product_link = 'https://www.bukalapak.com' + str(product_name.get('href'))
    product_name = product_name.text.replace('\n', "").strip() 
    d['product_link'] = product_link
    d['product_name'] = product_name
    l.append(d)

```

# Function
The following is a function that does all of above.

```python
def scrape():
    base_url = "https://www.bukalapak.com/products?utf8=%E2%9C%93&source=navbar&from=omnisearch&search_source=omnisearch_keyword&search%5Bhashtag%5D=&search%5Bkeywords%5D=majalah+popular"

    r = requests.get(base_url)
    soup = BeautifulSoup(r.text, "html.parser")
    all_product = soup.find_all('div', class_="product-card")
    l = []
    
    for item in all_product:
        d = {}

        product_name = item.find("a", {"class":"product__name"}) 
        product_link = 'https://www.bukalapak.com' + str(product_name.get('href'))
        product_name = product_name.text.replace('\n', "").strip() 
        d['product_link'] = product_link
        d['product_name'] = product_name
        l.append(d)
    
    return l

scrape()
```
